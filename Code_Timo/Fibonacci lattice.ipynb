```python
import numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d as a3
from scipy.spatial import ConvexHull, SphericalVoronoi, KDTree
import scipy.spatial as sc
import matplotlib.patches as patches
import random
import copy
import scipy
import math
import matplotlib.colors as colors
import PIL

def sph_to_cart(phi,theta): #for r = 1
    return np.array([np.cos(phi)*np.sin(theta),np.sin(phi)*np.sin(theta),np.cos(theta)])

def cart_to_sph(x,y,z):
    return np.array([np.arctan2(y,x),np.arctan2(np.sqrt(x**2+y**2),z)])
```

```python
%%time
# create the fibonacci lattice
numpoints = 101 #if this is even, the number of generated points will be one higher. If this is odd, it is equal
numpoints = numpoints//2 #divided by 2 because fibonacci_sphere goes from -N to N

def fibonacci_sphere(N):
    points = []
    g_angle = math.pi * (math.sqrt(5.) + 1)  # golden angle in radians

    for i in range(-N,N+1): #2N+1 points
        z = (2*i / (2*N+1)) # z goes from - 1 + 1/(2N+1) to 1 - 1/(2N+1), 2N + 1 points
        radius = math.sqrt(1 - z * z)  # radius of circle at z

        phi = g_angle * i  # golden angle increment

        x = math.cos(phi) * radius
        y = math.sin(phi) * radius

        points.append((x, y, z))

    return np.array(points)
points = fibonacci_sphere(numpoints)

x = points[:,0]
y = points[:,1]
z = points[:,2]
theta = np.arctan2(np.sqrt(x**2+y**2),z)
phi = np.arctan2(y,x)

points_phi_theta = np.stack((phi,theta),axis=1)
```

```python
%%time
# divide the lattice into triangles and get the area of each facet (convex hull of the lattice)
chull = ConvexHull(points)

facets = points[chull.simplices]
mid_of_facet = np.mean(facets,axis=1) #the midpoint of each facet
r1 = facets[:,1]-facets[:,0] #two sides of the facet
r2 = facets[:,2]-facets[:,1]
area_tri = 0.5*np.linalg.norm(np.cross(r1,r2),axis=1) #the area of each facet
print('number of facets: ',len(facets))

xmid = mid_of_facet[:,0]
ymid = mid_of_facet[:,1]
zmid = mid_of_facet[:,2] 

thetamid = np.arctan2(np.sqrt(xmid**2+ymid**2),zmid)
phimid = np.arctan2(ymid,xmid)

#put the facets into spherical coordinates (only phi and theta)
facets_sph = np.empty_like(facets[:,:,:2]) #get correct shape
facets_sph[:,:,0] = np.arctan2(facets[:,:,1],facets[:,:,0]) #phi
facets_sph[:,:,1] = np.arctan2(np.sqrt(facets[:,:,0]**2+facets[:,:,1]**2),facets[:,:,2])  #theta

#pick out the triangles on the borders and add or subtract 2pi for the points over the edge
idx_to_delete = [] #indices to delete (poles)
for i in range(len(facets_sph)):
    facet = facets_sph[i]
    #phi border
    if max(facet[:,0])-min(facet[:,0]) > np.pi*3/2: 
        if phimid[i] > 0: # if the midpoint is on the right +2pi for left points 
            for j in range(len(facet[:,0])):
                if facet[j,0] < 0:
                    facet[j,0] += 2*np.pi
        if phimid[i] < 0: #if its on the left: phi - 2pi for the right points
            for j in range(len(facet[:,0])):
                if facet[j,0] > 0:
                    facet[j,0] += -2*np.pi
fig = plt.figure()
ax = fig.add_subplot(111,projection='mollweide')
plotfacets_sph = copy.copy(facets_sph)
plotfacets_sph[:,:,1] += -np.pi/2

# i = 0
# for facet in plotfacets_sph:
#     c=(i%5)/5
#     ax.add_patch(patches.Polygon(facet,fill = True, color = [c,c,0.7]))
#     i+=1
# plt.axis('off')
# plt.show()
```

```python
#Spherical voronoi diagram of the fibonacci points
radius = 1
center = np.array([0, 0, 0])
sv = SphericalVoronoi(points, radius, center)
area_vor = sv.calculate_areas()
facets_vor = []                                  #facets of the voronoi sphere
for i in range(len(sv.regions)):
    facets_vor.append(sv.vertices[sv.regions[i]])

# save the voronoi facets with area
name = 'voronoi_grid_n' + str(2*numpoints+1)
voronoi_grid = np.concatenate((points_phi_theta,area_vor[:,None]),axis=1)
np.savetxt(name,voronoi_grid,delimiter=',') #phi, theta, area
np.save('facets_vor_n'+str(round(2*numpoints+1))+'.npy',facets_vor,allow_pickle=True)
```

```python
%%time
#3d plot and mollweide of the voronoi facets
fig = plt.figure(figsize = [16,8])
ax1 = fig.add_subplot(121,projection='3d')
ax2 = fig.add_subplot(122,projection='mollweide')

plot_facets_vor = []
i=0
for facet in facets_vor:
    fac = a3.art3d.Poly3DCollection([facet])
    c = float(np.random.rand(1))
    fac.set_color(colors.rgb2hex((c,c,0.7)))
#     ax1.add_collection3d(fac)
    
    phi = np.arctan2(facet[:,1],facet[:,0])
    theta = np.arctan2(np.sqrt(facet[:,0]**2+facet[:,1]**2),facet[:,2]) - np.pi/2
    phimid = np.arctan2(points[i,1],points[i,0])
    thetamid = np.arctan2(np.sqrt(points[i,0]**2+points[i,1]**2),points[i,2]) - np.pi/2
    facet_sph = np.dstack((phi,theta)).reshape(len(phi),2)
    
    #deal with the facets over the borders by adding or subtracting 2pi to the points on the other side  
    if max(facet_sph[:,0])-min(facet_sph[:,0]) > np.pi*3/2: 
        if phimid > 0: # if the midpoint is on the right +2pi for left points 
            for j in range(len(facet_sph[:,0])):
                if facet_sph[j,0] < 0:
                    facet_sph[j,0] += 2*np.pi
        if phimid < 0: #if its on the left: phi - 2pi for the right points
            for j in range(len(facet_sph[:,0])):
                if facet_sph[j,0] > 0:
                    facet_sph[j,0] += -2*np.pi 
    facet_sph[:,1] = -facet_sph[:,1]
    plot_facets_vor.append(facet_sph)
#     ax2.add_patch(patches.Polygon(facet_sph[sc.ConvexHull(facet_sph).vertices],fill = True, color = [c,c,0.7]))
    i+=1  
    
ax1.set_xlim(-1,1)
ax1.set_ylim(-1,1)
ax1.set_zlim(-1,1)
# _ = ax1.set_xticks([])
# _ = ax1.set_yticks([])
# _ = ax1.set_zticks([])
ax2.axis('off')
ax1.set_title('3D plot of voronoi diagram of fibonacci lattice')
ax2.set_title('mollweide projection of voronoi diagram of fib lattice')
ax1.elev = 0
ax1.azim = 270 # change this or elev to see different parts on the sphere

ax1.set_xlabel('y')
ax1.set_ylabel('x')
ax1.set_zlabel('z')

plt.show() 

# save the voronoi facets with area
name = 'plot_facets_n' + str(2*numpoints+1) + '.npy'
np.save(name,plot_facets_vor,allow_pickle=True)
```
